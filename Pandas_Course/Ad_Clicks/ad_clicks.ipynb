3.
If the column ad_click_timestamp is not null, then someone actually clicked on the ad that was displayed.
Create a new column called is_click, which is True if ad_click_timestamp is not null and False otherwise.

```python
ad_clicks['is_click'] = ~ad_clicks.ad_click_timestamp.isnull()
ad_clicks
```

4.
We want to know the percent of people who clicked on ads from each utm_source.
Start by grouping by utm_source and is_click and counting the number of 
user_id‘s in each of those groups. Save your answer to the variable clicks_by_source.

```python
clicks_by_source = ad_clicks.groupby(['utm_source', 'is_click']).count().reset_index()
clicks_by_source
```

5. Now let’s pivot the data so that the columns are is_click 
(either True or False), the index is utm_source, and the values are user_id.
Save your results to the variable clicks_pivot.

```python
clicks_pivot = clicks_by_source\
   .pivot(index='utm_source',
          columns='is_click',
          values='user_id')\
   .reset_index()
clicks_pivot
```

6.
Create a new column in clicks_pivot called percent_clicked which is
 equal to the percent of users who clicked on the ad from each utm_source.
Was there a difference in click rates for each source?

```python
clicks_pivot['percent_clicked'] = clicks_pivot[True] / (clicks_pivot[True] + clicks_pivot[False])
clicks_pivot
```

7.

8.
Using the column is_click that we defined earlier, check to see 
if a greater percentage of users clicked on Ad A or Ad B.

```python
ad_df = ad_clicks.groupby(['experimental_group', 'is_click']).user_id.count().reset_index()
ad_df
```

```python
ad_clicks_pivot = ad_df\
   .pivot(index='experimental_group',
          columns='is_click',
          values='user_id')\
   .reset_index()
ad_clicks_pivot
```

```python
ad_clicks_pivot['percent_clicked'] = ad_clicks_pivot[True] / (ad_clicks_pivot[True] + ad_clicks_pivot[False])
ad_clicks_pivot
```

9.
The Product Manager for the A/B test thinks that the clicks might have changed by day of the week.
Start by creating two DataFrames: a_clicks and b_clicks,
which contain only the results for A group and B group, respectively.

```python
a_clicks = ad_clicks[ad_clicks.experimental_group == 'A' ]
a_clicks
```

```python
b_clicks = ad_clicks[ad_clicks.experimental_group == 'B' ]
b_clicks
```

10.
For each group (a_clicks and b_clicks), calculate the percent of users who clicked on the ad by day.

```python
a_clicks_grouped = a_clicks.groupby(['day', 'is_click']).count().reset_index()
a_clicks_grouped
```

```python
b_clicks_grouped = b_clicks.groupby(['day', 'is_click']).count().reset_index()
b_clicks_grouped 
```

```python
a_clicks_pivot = a_clicks_grouped\
   .pivot(index='day',
          columns='is_click',
          values='user_id')\
   .reset_index()
a_clicks_pivot
```

```python
b_clicks_pivot = b_clicks_grouped\
   .pivot(index='day',
          columns='is_click',
          values='user_id')\
   .reset_index()
b_clicks_pivot
```

```python
a_clicks_pivot['percent_clicked'] = a_clicks_pivot[True] / (a_clicks_pivot[True] + a_clicks_pivot[False])
a_clicks_pivot
```

```python
b_clicks_pivot['percent_clicked'] = b_clicks_pivot[True] / (b_clicks_pivot[True] + b_clicks_pivot[False])
b_clicks_pivot
```
